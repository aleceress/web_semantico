```python
from possibilearn.fuzzifiers import *
import matplotlib
import matplotlib.pyplot as plt
matplotlib.use("pgf")
matplotlib.rcParams.update({
    "pgf.texsystem": "pdflatex",
    'font.family': 'serif',
    'text.usetex': True,
    'pgf.rcfonts': False,
})
fig = plt.figure()
fig.set_size_inches(w= 5.39749, h=3.5)
c = LinearFuzzifier()
cf = c.get_fuzzified_membership(3, [1,2,1,2,3,3,3,3,2,4,7,3,2], lambda x: x)
x = np.arange(0, 10, .01)
plt.plot(x, list(map(cf, x)))
plt.ylim(0, 1.1)
plt.xlim(0)
plt.savefig('linearFuzzifier.pgf')
```
